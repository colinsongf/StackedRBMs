```python
import Image

def tiles(height, width, images, imwidth, imheight, welastic=False, helastic=False, figsize=(10, 10)):
    f = figure(figsize=figsize)
    ax = f.add_subplot(111)
    for i in xrange(height):
        for j in xrange(width):
            id = (i - 1)*width + j
            im = images[id]
            _imwidth = im.size/imheight if welastic else imwidth
            _imheight = im.size/imwidth if helastic else imheight
            assert _imwidth*_imheight==im.size, "%d %d != %d" % (_imwidth, _imheight, im.size)
            im = im.reshape(_imheight, _imwidth)
            ax.imshow(im, 
                      extent=numpy.array([j*imwidth, j*imwidth +_imwidth, 
                                          i*imheight, i*imheight + _imheight]), 
                      cmap=gray())

    plt.xlim(0,imwidth * width)
    plt.ylim(0,imheight * height)
```

```python
def replace_nnp(sent, by="entity"):
    
    def parse_tree(tree):
        result = ""
        for e in tree:
            if type(e) is nltk.tree.Tree:
                result += parse_tree(e)
            elif e[1] == "NNP":
                result += " " + by
            else:
                result += " " + e[0]
        return result
        
    words = nltk.word_tokenize(sent)
    pos = nltk.pos_tag(words)
    tree = nltk.ne_chunk(pos, True)
    return parse_tree(tree)
```

```python
data_home = '/home/trung/phd/datasets/sub_obj/sub_obj_sents.1.2.tsv'
def spatial_pyramid_features(x, level):
    ''' spatial pyramid features of x, where x is a list of 2 
    dimensional binary images. Refer to this document for more 
    information http://www.cs.unc.edu/~lazebnik/publications/pyramid_chapter.pdf.'''
    result = [1.0/(2**level)*x.sum().sum()]
    for i in xrange(0, level):
        for j in xrange(x.shape[0]):
            rows, cols = x[j].shape
            cell_row = max(rows*1./(2**i), 1)
            cell_col = max(cols*1./(2**i), 1)
            for r in xrange(2**i):
                for c in xrange(2**i):
                    w = 1.0/(2**(level - i))
                    rmask = slice(int(round(r*cell_row)), int(round((r + 1)*cell_row)))
                    cmask = slice(int(round(c*cell_col)), int(round((c + 1)*cell_col)))
                    result.append(w*x[j][rmask, cmask].sum())
    return result                            

def spatial_pyramid_kernel(x, y):
    return asarray([x, y]).min(axis=0).sum()

f = open(data_home)
reader = csv.reader(f, delimiter='\t')
X0, Y0 = [], []
for sent, label in reader:
    _ = []
    """
    for w in padded(nltk.word_tokenize(replace_nnp(sent))):
        _.append(google_embeddings.vector(w).flatten())
    X0.append(asarray(_)/stds)
    """
    Y0.append(label)
f.close()

X0 = asarray(X0, dtype=object).reshape(len(X0), 1)
Y0 = [{'objective': 0, 'subjective': 1}.get(y) for y in Y0]
Y0 = asarray(Y0)
```
