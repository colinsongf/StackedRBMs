# Model classes

```python
# Define Binary Tree structure
class BTree(object):
    
    def __init__(self, meta=None):
        self.meta, self.parent, self.left, self.right =\
            meta, None, None, None
            
    def add_child(self, child):
        if self.left is None:
            self.left = child
        else:
            self.right = child
        child.parent = self

    def iternodes(self):
        for leaf in [self.left, self.right]:
            if leaf is not None:
                for node in leaf.iternodes():
                    yield node
        yield self
        
    def is_leaf(self):
        return self.left is None and self.right is None
                    
class PhraseBTree(BTree):
    
    def add_child(self, child):
        if self.left is not None and self.right is not None:
            raise ValueError("Tree is full")           
        if self.left is None:
            self.left = child
            self.meta["order"] = child.meta["order"]
        else:
            if self.left.meta["order"] > child.meta["order"]:
                self.right = self.left
                self.left = child
            else:
                self.right = child
            self.meta["order"] = self.right.meta["order"]
        child.parent = self                    
```

```python
class RNTN(object):
    
    def __init__(self, dim, dict_size, classes):
        self.dim = dim
        self.x = tn.shared(rand(dict_size, dim, 1)*.01)
        self.V = tn.shared(rand(dim, 2*dim, 2*dim)*.01)
        self.W = tn.shared(rand(dim, 2*dim)*.01)
        self.Ws = tn.shared(rand(classes, dim)*.01)
        
    def kernel(self, a, b):
        ab = T.concatenate([a, b])
        return T.dot(T.dot(ab.T, self.V), ab).T.reshape((self.dim, 1)) + T.dot(self.W, ab)
    
    @staticmethod
    def softmax(x):
        return T.exp(x)/T.sum(T.exp(x))
    
    @staticmethod
    def KL(p1, p2):
        return T.sum(p1*T.log(p1/p2))

    def L2(self):
        return T.sum(self.x**2) + T.sum(self.V**2) +\
               T.sum(self.W**2) + T.sum(self.Ws**2)
    
    def _cost(self, tree):
        for node in tree.iternodes():
            if node.left is None:
                node.meta["value"] = self.x[node.meta["windex"]]
            else:
                left_value = node.left.meta["value"]
                right_value = node.right.meta["value"]
                node.meta["value"] = self.kernel(left_value, right_value)
                                   
        return T.sum([RNTN.KL(
                        asarray([[1 - node.meta["score"], node.meta["score"]]]).T,
                        RNTN.softmax(T.dot(self.Ws, node.meta["value"])), 
                      ) for node in tree.iternodes()])
    
    def cost(self, trees):
        return T.sum([self._cost(tree) for tree in trees])
    
    def _train(self, tree_batch, lr=.01):
        cost = self.cost(tree_batch)
        gx = T.grad(cost, self.x)
        gV = T.grad(cost, self.V)
        gW = T.grad(cost, self.W)
        gWs = T.grad(cost, self.Ws)
        updates = [
            (self.x, self.x - lr*gx),
            (self.V, self.V - lr*gV),
            (self.W, self.W - lr*gW),
            (self.Ws, self.Ws - lr*gWs)
            ]
        trainf = tn.function(inputs=[], outputs=cost, updates=updates)
        return trainf()
        
    def train(self, trees, lr=.01, epochs=100, verbose=True, batch=10):
        trees = asarray(trees)
        batches = int(ceil(trees.shape[0]*1./batch))
        costs = []
        for epoch in xrange(epochs):
            print "Training epoch %d... " % epoch
            permutation = random.permutation(trees.shape[0])
            _costs = []
            for bid in xrange(batches):
                if verbose:
                    print "Batch %d... " % bid,
                cost = self._train(trees[bid*batch:(bid + 1)*batch], lr)
                if verbose:
                    print cost
                _costs.append(cost)
            mcost = asarray(_costs).mean()
            print mcost
            costs.append(mcost)
        return costs            
```

# Load data

```python
# Load phrase dictionary
phrases = {}
fphrase = open(os.path.join(DATA_PATH, "dictionary.txt"))
for line in fphrase:
    phrase, index = line.split("|")
    phrases[phrase] = {"index": int(index)}
fphrase.close()

# Load phrase sentiments
phraseid_score = {}
flabel = open(os.path.join(DATA_PATH, "sentiment_labels.txt"))
flabel.next()
for line in flabel:
    index, score = line.split("|")
    phraseid_score[int(index)] = float(score)
flabel.close()
for p in phrases:
    phrases[p]["score"] = phraseid_score[phrases[p]["index"]]
    
fwords = open(os.path.join(DATA_PATH, "SOStr.txt"))
ftree = open(os.path.join(DATA_PATH, "STree.txt"))
trees = []
wordSet = set()
for wordStr, treeStr in izip(fwords, ftree):
    words = wordStr.strip().split("|")
    for word in words:
        wordSet.add(word)
    parentIds = treeStr.split("|")
    nodes = {}
    for i, w in enumerate(words, 1):
        nodes[i] = PhraseBTree({"word": w, "order": i})
    for i, pid in enumerate(parentIds, 1):
        pid = int(pid)
        if pid==0:
            trees.append(nodes[i])
        else:
            if pid not in nodes:
                nodes[pid] = PhraseBTree({"word": None, "order": pid})
            nodes[pid].add_child(nodes[i])
fwords.close()
ftree.close()
words = {word:i for i, word in enumerate(wordSet)}
```

```python
# Fill tree with index and label
def fillTree(tree, phraseDict):
    for node in tree.iternodes():
        if node.meta["word"] is None:
            node.meta["word"] = node.left.meta["word"] + " " + node.right.meta["word"]
        node.meta["index"] = phraseDict[node.meta["word"]]["index"]
        node.meta["score"] = phraseDict[node.meta["word"]]["score"]
        node.meta["windex"] = words.get(node.meta["word"])

for tree in trees:
    fillTree(tree, phrases)
```
